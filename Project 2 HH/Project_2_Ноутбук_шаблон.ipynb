```python
import requests
from bs4 import BeautifulSoup as BS
# request for parsing
headers = {'User-Agent': 'Mozilla/5.0'}
url = 'https://ru.wikipedia.org/wiki/Города-миллионеры_России'
response = requests.get(url, headers=headers)
page = BS(response.text, 'html.parser') # get HTML-page
cities = []
for row in page.find_all('tr'):
    cells = row.find_all('td')
    if len(cells) >= 2:
        num_cell = cells[0].get_text(strip=True)
        if num_cell.isdigit() and 1 <= int(num_cell) <= 16:
            city_link = cells[1].find('a')
            if city_link:
                city = city_link.get_text(strip=True)
                cities.append(city)

cities_tuple = "', '".join(cities)
```

```python
query_5_7 = f'''
SELECT 
    a.name AS region_name,
    COUNT(v.id) AS vacancies_count
FROM vacancies v
INNER JOIN employers e ON v.employer_id = e.id
INNER JOIN areas a ON v.area_id = a.id
WHERE e.name = 'Яндекс'
  AND a.name IN ('{cities_tuple}')
GROUP BY a.name

UNION ALL

SELECT 
    'Total' AS region_name,
    COUNT(v.id) AS vacancies_count
FROM vacancies v
INNER JOIN employers e ON v.employer_id = e.id
WHERE e.name = 'Яндекс'
  AND v.area_id IN (SELECT id FROM areas WHERE name IN ('{cities_tuple}'))
'''
```

6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в **среднем** указывают для каждого типа требуемого опыта (уникальное значение из поля *experience*). 

При решении задачи примите во внимание следующее:
1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
3. Если в расчётах участвует null, в результате он тоже даст null (посмотрите, что возвращает запрос select 1 + null). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит null на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до целого числа. 

```python
query_6_6 = f'''
SELECT 
    experience,
    ROUND(AVG(
        (COALESCE(salary_from, 0) + COALESCE(salary_to, 0)) / 
        CASE 
            WHEN salary_from IS NOT NULL AND salary_to IS NOT NULL THEN 2
            WHEN salary_from IS NOT NULL THEN 1
            WHEN salary_to IS NOT NULL THEN 1
            ELSE 0
        END
    )) AS avg_salary
FROM vacancies
WHERE (
    LOWER(name) LIKE '%data scientist%'
    OR LOWER(name) LIKE '%data science%'
    OR LOWER(name) LIKE '%исследователь данных%'
    OR LOWER(name) LIKE '%machine learning%'
    OR LOWER(name) LIKE '%машинн%обучен%'
    OR LOWER(name) LIKE '%дата сайентист%'
    OR (name LIKE '%ML%' AND name NOT ILIKE '%HTML%')
)
AND LOWER(name) NOT LIKE '%html%'
AND LOWER(name) NOT LIKE '%xml%'
AND (salary_from IS NOT NULL OR salary_to IS NOT NULL)
GROUP BY experience
ORDER BY experience
'''
```

```python
pd.read_sql(query_6_6, connection)
```
