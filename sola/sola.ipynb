```python
import pandas as pd
```

```python
solar_data = pd.read_csv('C:\IDE\sola\solarpower.csv')
solar_data.head()
```

```python
# Заменяем -1 на NaN
solar_data['cum_power'] = solar_data['cum_power'].replace(-1, pd.NA)
# Линейная интерполяция между соседними значениями
solar_data['cum_power'] = solar_data['cum_power'].interpolate(method='linear')
# Среднее после заполнения пропусков
mean_cum_power = round(solar_data['cum_power'].mean())
print(mean_cum_power)
```
