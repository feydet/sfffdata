```python
def calculate_dcf(cf, rates):
    discounted_flows = []
    discount_factor = 1

    for i in range(len(cf)):
        discount_factor *= (1 + rates[i])
        dcf = cf[i] / discount_factor
        discounted_flows.append(dcf)

    return discounted_flows


# ДАННЫЕ
cash_flows = [-431000, 1830000, 2040000]
rates = [0.18, 0.20, 0.22]

# РАСЧЁТ
dcf_values = calculate_dcf(cash_flows, rates)

# ВЫВОД
total_dcf = 0
for i in range(len(dcf_values)):
    total_dcf += dcf_values[i]
    print(f"Год {i+1}: ДДП = {dcf_values[i]:.2f} руб.")

print(f"\nСуммарный DCF (NPV): {total_dcf:.2f} руб.")
```

```python
def calculate_discounted_flows(cf, rates):
    discounted = []
    discount_factor = 1

    for i in range(len(cf)):
        discount_factor *= (1 + rates[i])
        dcf = cf[i] / discount_factor
        discounted.append(dcf)

    return discounted


def dynamic_payback_period(initial_investment, cf, rates):
    # считаем дисконтированные потоки
    discounted_flows = calculate_discounted_flows(cf, rates)

    # берём только положительные потоки (доходы)
    positive_flows = [x for x in discounted_flows if x > 0]

    # средний дисконтированный поток
    avg_discounted_cf = sum(positive_flows) / len(positive_flows)

    # срок окупаемости
    payback = initial_investment / avg_discounted_cf

    return payback


# ДАННЫЕ
initial_investment = 431000
cash_flows = [-431000, 1830000, 2040000]
rates = [0.18, 0.20, 0.22]

# РАСЧЁТ
payback = dynamic_payback_period(initial_investment, cash_flows, rates)

print(f"Срок окупаемости (динамический): {payback:.2f} года")
```
